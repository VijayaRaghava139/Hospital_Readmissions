```python
# Patient_ID, Num_Lab_Procedures, Num_Diagnoses columns are not related to the Target column.

df = df.drop(columns=['Patient_ID'])
```

```python
df = df.drop_duplicates()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X = df.drop(columns=['Readmitted'])
y = df['Readmitted']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
def Detect_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_value = Q1 - (1.5 * IQR)
    upper_value = Q3 + (1.5 * IQR)
    return lower_value, upper_value 
```

```python
lower_age, upper_age = Detect_outliers(X_train['Age'])
X_train[(X_train['Age'] <= lower_age) | (X_train['Age'] >= upper_age)]
```

```python
X_train['Age'] = X_train['Age'].apply(lambda x: X_train['Age'].median() if ((x <= lower_age) | (x >= upper_age)) else x)
```

```python
lower_medications, upper_medications = Detect_outliers(X_train['Num_Emergency_Visits'])
X_train[(X_train['Num_Emergency_Visits'] <= lower_medications) | (X_train['Num_Emergency_Visits'] >= upper_medications)]
```

```python
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder, LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
le = LabelEncoder()
```

```python
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

```python
# def random_impute(column):
#     if isinstance(column, pd.DataFrame):
#         column = column.iloc[:, 0]
#     categories = column.dropna().unique() 
#     return column.apply(lambda x: np.random.choice(categories) if pd.isna(x) else x)

# random_imputer = FunctionTransformer(random_impute, validate=False)
```

```python
categories = X_train['A1C_Result'].dropna().unique() 

X_train['A1C_Result'] = X_train['A1C_Result'].apply(lambda x: np.random.choice(categories) if pd.isna(x) else x)
X_test['A1C_Result'] = X_test['A1C_Result'].apply(lambda x: np.random.choice(categories) if pd.isna(x) else x)

# X_train = X_train.drop(columns=['A1C_Result'])
# X_test = X_test.drop(columns=['A1C_Result'])
```

```python
ohe_cols = ['Gender', 'Admission_Type', 'Diagnosis', 'A1C_Result']
standardScaler_cols = [col for col in X_train.select_dtypes(include='int64').columns]
standardScaler_cols
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('ohe', OneHotEncoder(handle_unknown='ignore'), ohe_cols),
        ('standardscaler', StandardScaler(), standardScaler_cols)
    ],
)
```

```python
pipeline_lr = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('lr', LogisticRegression())
    ]
)
```

```python
pipeline_lr.fit(X_train, y_train)
```

```python
y_pred_lr = pipeline_lr.predict(X_test)
```

```python
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f'Accuracy is : {accuracy_lr}')
```

```python
pipeline_rf = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('rf', RandomForestClassifier(random_state=42))
    ]
)

pipeline_rf.fit(X_train, y_train)
```

```python
y_pred_rf = pipeline_rf.predict(X_test)

accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f'Accuracy is : {accuracy_rf}')
```

```python
param_grid_rf = {
    'rf__n_estimators' : [50,75,100,150,200,300,500],
    'rf__criterion' : ['gini', 'entropy', 'log_loss'],
    'rf__max_depth': [None, 10, 20, 30],
    'rf__min_samples_split': [2, 5, 10],
    'rf__max_features' :[None, 'sqrt', 'log2'],
    'rf__min_samples_leaf': [1, 2, 4]
}
```

```python
rf = RandomForestClassifier(random_state=42)

grid_search = GridSearchCV(estimator=pipeline_rf, param_grid=param_grid_rf, cv=5, verbose=2, n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_pipeline = grid_search.best_estimator_
y_pred_rf = best_pipeline.predict(X_test)

print(classification_report(y_test, y_pred_rf))
```

```python
accuracy_score(y_test, y_pred_rf)
```

```python
best_params = grid_search.best_params_

print(f"Best Hyperparameters: {best_params}")
```
